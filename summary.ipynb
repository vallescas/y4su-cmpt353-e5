```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
def sumstats(filename: str, title: str) -> None:
    """
    Provides summary statistics of a DataFrame.

    Args: 
        filename (str): The name of the csv file. 
        title (str): The title for the scatterplot figure.

    Returns:
        None: This function does not return any value. Instead, it prints a string that displays the 
                mean, std, range (min/max), and correlation coefficient of the DataFrame.
    """

    # reading the data
    data = pd.read_csv(filename)

    # displaying the df
    display(data)

    # creating a scatterplot
    plt.plot(data['x'], data['y'], 'b.', alpha=0.5)
    plt.xlabel('x-values')
    plt.ylabel('y-values')
    plt.title(title)
    plt.show()
    
    # the mean of both x and y axes
    # -- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.mean.html
    print('====================== mean =======================')
    print('The mean across the x-axis is: ' + str(data.mean().x))
    print('The mean across the y-axis is: ' + str(data.mean().y))
    
    # the standard deviation of both x and y axes
    # -- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.std.html
    print('====================== std ========================')
    print('The std across the x-axis is: ' + str(data.std().x))
    print('The std across the y-axis is: ' + str(data.std().y))
    
    # the minimum value in the x and y axes
    # -- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.min.html
    print('====================== min ========================')
    print('The min across the x-axis is: ' + str(data.min().x))
    print('The min across the y-axis is: ' + str(data.min().y))
    
    # the maximum value in the x and y axes
    # -- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.max.html
    print('====================== max ========================')
    print('The max across the x-axis is: ' + str(data.max().x))
    print('The max across the y-axis is: ' + str(data.max().y))
    
    # the correlation coefficient of the x and y axes
    # -- https://ggbaker.ca/data-science/content/stats.html
    print('====================== cc =========================')
    print('The correlation coefficient is: ' + str(stats.linregress(data['x'], data['y']).rvalue))
```

```python
sumstats(filename='data-1.csv', title='data-1.csv')
```
